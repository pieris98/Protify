```python
import pandas as pd
```

```python
df = pd.read_csv('OMT.csv')
df.head()
```

```python
df = df.rename(columns={'protein seq': 'seqs', 'normalized growthrate': 'labels'}, inplace=False)
df2 = df[['seqs', 'labels']]
df2.head()
```

```python
from datasets import Dataset, DatasetDict


# randomly split into 0.8, 0.1, 0.1 train, valid, test


df2 = df2.sample(frac=1).reset_index(drop=True)
train = df2.iloc[:int(0.8*len(df2))]
valid = df2.iloc[int(0.8*len(df2)):int(0.9*len(df2))]
test = df2.iloc[int(0.9*len(df2)):]


dataset = DatasetDict({
    'train': Dataset.from_pandas(train),
    'valid': Dataset.from_pandas(valid),
    'test': Dataset.from_pandas(test)
})

dataset
```

```python
dataset.push_to_hub('GleghornLab/millionfull_round_1_oct_2025')
```
